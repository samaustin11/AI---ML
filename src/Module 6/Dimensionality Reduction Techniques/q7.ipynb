```python
# Question 7: Combining PCA and t-SNE for Better Results
# Description: Combine PCA and t-SNE for large datasets â€“ reduce dimensions using PCA, visualize with t-SNE.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target

# Standardize the data (important for PCA and t-SNE)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 1: Apply PCA to reduce the dimensionality
pca = PCA(n_components=2)  # Set n_components to 2 or a number <= 4
X_pca = pca.fit_transform(X_scaled)

# Step 2: Apply t-SNE to the PCA-reduced data
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_pca)

# Step 3: Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=50)
plt.colorbar(label='Target Class')
plt.title('t-SNE Visualization of PCA-Reduced Data')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```
