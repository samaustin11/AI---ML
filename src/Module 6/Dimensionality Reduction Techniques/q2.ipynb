```python
# Question 2: Geometry of PCA
# Description: Compute PCA on a simple 2D matrix and manually derive its principal components.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a simple 2D dataset (2 features, 5 samples)
X = np.array([[2.5, 3.5],
              [3.5, 4.5],
              [3.0, 3.0],
              [4.0, 5.0],
              [4.5, 4.0]])

# Step 2: Standardize the data (mean centering)
mean = np.mean(X, axis=0)
X_centered = X - mean

# Step 3: Compute the covariance matrix
cov_matrix = np.cov(X_centered, rowvar=False)

# Step 4: Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Step 5: Sort eigenvalues and eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]  # Sort in descending order
eigenvalues_sorted = eigenvalues[sorted_indices]
eigenvectors_sorted = eigenvectors[:, sorted_indices]

# Step 6: The principal components are the eigenvectors corresponding to the largest eigenvalues
pc1 = eigenvectors_sorted[:, 0]
pc2 = eigenvectors_sorted[:, 1]


# Step 7: Project the data onto the principal components
X_projected = X_centered.dot(eigenvectors_sorted)

# Step 8: Plot the original data and the principal components
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], color='blue', label='Original Data')
plt.quiver(mean[0], mean[1], pc1[0], pc1[1], angles='xy', scale_units='xy', scale=1, color='red', label='PC1')
plt.quiver(mean[0], mean[1], pc2[0], pc2[1], angles='xy', scale_units='xy', scale=1, color='green', label='PC2')
plt.title("PCA: Geometry of Principal Components")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()

# Display the eigenvalues (variance explained by each principal component)
print(f"Eigenvalues (Variance Explained): {eigenvalues_sorted}")
```
