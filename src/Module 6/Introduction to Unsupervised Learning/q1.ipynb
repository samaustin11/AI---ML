```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml
from sklearn import datasets
from sklearn.datasets import make_classification
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_sample_image

# ---- Task 1: Identifying patterns in customer purchasing behavior ----

# Simulating a customer purchasing dataset (e.g., Annual Income and Spending Score)
np.random.seed(42)
customers = pd.DataFrame({
    'Annual Income (k$)': np.random.randint(20, 120, 500),
    'Spending Score (1-100)': np.random.randint(1, 100, 500)
})
# Standardizing the features
scaler = StandardScaler()
X = scaler.fit_transform(customers)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
customers['Cluster'] = kmeans.fit_predict(X)

# Visualizing customer segments
plt.figure(figsize=(8,6))
plt.scatter(customers['Annual Income (k$)'], customers['Spending Score (1-100)'], c=customers['Cluster'], cmap='viridis')
plt.title("Customer Segments based on Income & Spending Score")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.colorbar()
plt.show()
```

```python
# ---- Task 2: Grouping news articles by topic ----

# Simulate a collection of news articles (e.g., topic-related content)
documents = [
    "Stock market hits new high despite economic uncertainty",
        "New technology revolutionizes the healthcare industry",
            "Scientists discover new treatment for heart disease",
                "The latest trends in global technology startups",
                    "A new model for understanding climate change"
                    ]

# Vectorizing the text data using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# Apply KMeans to group articles into clusters (topics)
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X)

# Display the articles with their respective clusters
for i, label in enumerate(labels):
    print(f"Cluster {label}: {documents[i]}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.datasets._base import load_sample_image  # Correct import

# Load a sample image (China image from sklearn)
china = load_sample_image("china.jpg")

# Reshape image: each pixel is a data point with 3 features (RGB)
X = china.reshape((-1, 3))

# Standardize the pixel values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.6, min_samples=50)
labels = dbscan.fit_predict(X_scaled)

# Reshape labels to image dimensions
label_image = labels.reshape(china.shape[:2])

# Visualize clustering results
plt.figure(figsize=(10, 6))
plt.imshow(label_image, cmap='tab20')
plt.title("Land Use Clusters Identified via DBSCAN")
plt.axis('off')
plt.colorbar()
plt.show()
```
