```python
# Question 3: Implementing DBSCAN Clustering
# Description: Perform DBSCAN clustering on a dataset with noise and visualize the clusters.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate synthetic dataset with noise (some random data points as noise)
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
X = np.vstack([X, np.random.uniform(-10, 10, size=(50, 2))])  # Add noise

# Perform DBSCAN clustering
dbscan = DBSCAN(eps=1, min_samples=5)
y_pred = dbscan.fit_predict(X)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', marker='o')

# Mark noise points as red
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], c='red', label='Noise', marker='x')

plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
