```python
# Question 8: Parameter Tuning in DBSCAN
# Description: Experiment with different values of eps and min_samples in DBSCAN and observe the effect on clustering results.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate synthetic dataset with 2 features
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
X = np.vstack([X, np.random.uniform(-10, 10, size=(50, 2))])  # Add noise

# Experiment with different values of eps and min_samples
eps_values = [0.5, 1.0, 1.5]
min_samples_values = [3, 5, 10]

plt.figure(figsize=(12, 12))

for i, eps in enumerate(eps_values):
        for j, min_samples in enumerate(min_samples_values):
            # Apply DBSCAN with current eps and min_samples
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            y_pred = dbscan.fit_predict(X)

            # Plot the clustering result
            plt.subplot(len(eps_values), len(min_samples_values), i * len(min_samples_values) + j + 1)
            plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', marker='o')
            plt.title(f'eps={eps}, min_samples={min_samples}')
            plt.xlabel('Feature 1')
            plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()


```
