```python
# Title: Perform PCA for Dimensionality Reduction

# Task 1: Reducing Dimensionality of a Breast Cancer Dataset
# Objective: Reduce dimensions for visualization or further analysis.
# Dataset: Use the "Breast Cancer Wisconsin" dataset available in Scikit-learn.

# Task 2: Dimensionality Reduction for MNIST Dataset
# Objective: Visualize high-dimensional image data in a 2D space.
# Dataset: Use the MNIST handwritten digits dataset.

# Task 3: Reducing Dimensionality of Wine Quality Data
# Objective: Simplify a complex dataset for exploratory analysis.
# Dataset: Use the "Wine Quality" dataset from UCI Machine Learning Repository.
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Create synthetic dataset with 30 features (simulating the breast cancer dataset)
X, y = make_classification(n_samples=500, n_features=30, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to 2 components for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot the reduced dimensions
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', alpha=0.7)
plt.title('PCA of Synthetic Breast Cancer-like Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Class')
plt.show()
```

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA

# Load MNIST dataset from OpenML
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist["data"], mnist["target"]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to 2 components for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot the reduced dimensions
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y.astype(int), cmap='viridis', edgecolor='k', alpha=0.7)
plt.title('PCA of MNIST Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Digit Class')
plt.show()
```

```python
from sklearn.datasets import make_regression
from sklearn.decomposition import PCA

# Simulate wine quality dataset with 13 features
X, y = make_regression(n_samples=500, n_features=13, noise=0.1, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot the reduced dimensions
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', alpha=0.7)
plt.title('PCA of Simulated Wine Quality-like Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Wine Quality')
plt.show()
```
