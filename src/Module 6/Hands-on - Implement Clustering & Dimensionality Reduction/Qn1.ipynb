```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Create synthetic dataset for customer purchases (Annual Income and Spending Score)
data = {
    'CustomerID': range(1, 101),
        'AnnualIncome': np.random.randint(20, 120, size=100) * 1000,
            'SpendingScore': np.random.randint(1, 101, size=100)
            }

df = pd.DataFrame(data)

# Features for clustering
X = df[['AnnualIncome', 'SpendingScore']]

# Perform K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['AnnualIncome'], df['SpendingScore'], c=df['Cluster'], cmap='viridis')
plt.title('Customer Segmentation using K-Means Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
X_iris = iris.data
y_iris = iris.target

# Perform K-Means clustering
kmeans_iris = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans_iris.fit_predict(X_iris)

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=y_kmeans, cmap='viridis')
plt.title('Iris Dataset Clustering using K-Means')
plt.xlabel('Feature 1 (Sepal Length)')
plt.ylabel('Feature 2 (Sepal Width)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.datasets import load_sample_image
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Load the sample image (ensure it's available in the environment)
china = load_sample_image("china.jpg")

# Reshape the image to a 2D array of pixels (each pixel is represented by RGB values)
image_data = china.reshape((-1, 3))

# Perform K-Means clustering to reduce the number of colors (e.g., reduce to 16 colors)
kmeans_image = KMeans(n_clusters=16, random_state=42)
labels = kmeans_image.fit_predict(image_data)

# Replace each pixel's color by the centroid of the cluster it belongs to
centroids = kmeans_image.cluster_centers_
compressed_image = centroids[labels].reshape(china.shape)

# Display the original and compressed images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plot original image
ax[0].imshow(china)
ax[0].set_title('Original Image')
ax[0].axis('off')  # Turn off axis for cleaner display

# Plot compressed image
ax[1].imshow(compressed_image)
ax[1].set_title('Compressed Image (K-Means)')
ax[1].axis('off')  # Turn off axis for cleaner display

# Save the output to a file (if display fails)
plt.tight_layout()
plt.savefig("image_compressed_output.png")

# Show the plot (for environments that support it)
plt.show()

# Print message to indicate the saved file
print("Saved the compressed image as 'image_compressed_output.png'. Please check the file.")
```
