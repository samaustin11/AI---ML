Task 1:<br>
Objective: Identify if an email is spam or not spam.<br>
Load the UCI Spambase Dataset.<br>
Goal: Create a model that classifies emails into two categories: "spam" and "not spam."


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Create a synthetic dataset for spam detection
# Features will be word frequencies (simulated), and the target will be 1 for spam, 0 for not spam
np.random.seed(42)

# Simulate 1000 emails with 5 features (word frequencies)
data_size = 1000
feature_1 = np.random.rand(data_size)  # Feature 1: Word frequency of "free"
feature_2 = np.random.rand(data_size)  # Feature 2: Word frequency of "win"
feature_3 = np.random.rand(data_size)  # Feature 3: Word frequency of "offer"
feature_4 = np.random.rand(data_size)  # Feature 4: Word frequency of "buy"
feature_5 = np.random.rand(data_size)  # Feature 5: Word frequency of "limited"
labels = np.random.choice([0, 1], size=data_size)  # 0 for not spam, 1 for spam

# Create a DataFrame
data = pd.DataFrame({
    'feature_1': feature_1,
        'feature_2': feature_2,
            'feature_3': feature_3,
            'feature_4': feature_4,
                'feature_5': feature_5,
                    'label': labels
                    })
# Split features and target
X = data.drop('label', axis=1)
y = data['label']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train a RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Plotting the Confusion Matrix
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Spam', 'Spam'], yticklabels=['Not Spam', 'Spam'])
plt.title("Confusion Matrix for Spam Classification (Random Forest)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

Task 2:<br>
Objective: Diagnose whether a tumor is malignant or benign.<br>
Load the Breast Cancer Wisconsin dataset.<br>
Goal: Build a binary classification model to classify tumors.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Breast Cancer Wisconsin dataset from sklearn
data = load_breast_cancer()

# Convert it into a DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target  # Target column: 0 for malignant, 1 for benign

# Split features and target
X = df.drop('target', axis=1)
y = df['target']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train a RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Plotting the Confusion Matrix
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Malignant', 'Benign'], yticklabels=['Malignant', 'Benign'])
plt.title("Confusion Matrix for Tumor Classification (Random Forest)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```


Task 3:<br>
Objective: Determine whether a transaction is fraudulent or legitimate.<br>
Use a credit card transaction dataset.<br>
Goal: Classify transactions into "fraudulent" and "legitimate" categories.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Create a synthetic dataset for credit card transactions
np.random.seed(42)

# Simulate 1000 transactions with 5 features
data_size = 1000

# Feature 1: Transaction amount (in USD)
feature_1 = np.random.uniform(5, 500, size=data_size)

# Feature 2: Time of transaction (hour of day)
feature_2 = np.random.randint(0, 24, size=data_size)

# Feature 3: Merchant category (1: electronics, 2: groceries, 3: clothing, 4: others)
feature_3 = np.random.choice([1, 2, 3, 4], size=data_size)

# Feature 4: User's age group (0: under 18, 1: 18-25, 2: 26-40, 3: 41-60, 4: 60+)
feature_4 = np.random.choice([0, 1, 2, 3, 4], size=data_size)

# Feature 5: Transaction location (1: local, 2: international)
feature_5 = np.random.choice([1, 2], size=data_size)

# Target label: 1 for fraudulent, 0 for legitimate
labels = np.random.choice([0, 1], size=data_size, p=[0.95, 0.05])  # 5% fraud, 95% legitimate

# Create a DataFrame
data = pd.DataFrame({
    'transaction_amount': feature_1,
        'transaction_time': feature_2,
            'merchant_category': feature_3,
                'user_age_group': feature_4,
                    'transaction_location': feature_5,
                        'fraudulent': labels
                        })
# Split features and target
X = data.drop('fraudulent', axis=1)
y = data['fraudulent']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train a RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Plotting the Confusion Matrix
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Legitimate', 'Fraudulent'], yticklabels=['Legitimate', 'Fraudulent'])
plt.title("Confusion Matrix for Transaction Classification (Random Forest)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
