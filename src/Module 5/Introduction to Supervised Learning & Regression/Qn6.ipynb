Title: Understanding Regression Metrics

Task 1: Calculate MAE and MSE on test predictions and compare errors.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Task 1: Calculate MAE and MSE on test predictions and compare errors.

# Generate synthetic data (e.g., relationship between X and y)
np.random.seed(42)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = 3 * X.ravel() + np.random.normal(0, 1, X.shape[0])  # Linear relationship with noise

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fit a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Print the errors
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Visualizing the model's predictions vs true values
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='True Values', alpha=0.7)
plt.plot(X_test, y_pred, color='red', label='Predictions')
plt.title(f'Linear Regression: MAE = {mae:.2f}, MSE = {mse:.2f}')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

Task 2: Evaluate R2 Score on varying datasets and discuss significance.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Task 2: Evaluate R2 Score on varying datasets and discuss significance.

# Generate synthetic datasets with different complexities

# Simple linear relationship
X_simple = np.linspace(0, 10, 100).reshape(-1, 1)
y_simple = 2 * X_simple.ravel() + 3 + np.random.normal(0, 1, X_simple.shape[0])

# Non-linear relationship
X_non_linear = np.linspace(0, 10, 100).reshape(-1, 1)
y_non_linear = np.sin(X_non_linear.ravel()) * 10 + np.random.normal(0, 1, X_non_linear.shape[0])

# Random noise data
X_noise = np.random.rand(100, 1) * 10
y_noise = np.random.randn(100) * 10

# Split all datasets into training and testing sets
X_train_simple, X_test_simple, y_train_simple, y_test_simple = train_test_split(X_simple, y_simple, test_size=0.3, random_state=42)
X_train_non_linear, X_test_non_linear, y_train_non_linear, y_test_non_linear = train_test_split(X_non_linear, y_non_linear, test_size=0.3, random_state=42)
X_train_noise, X_test_noise, y_train_noise, y_test_noise = train_test_split(X_noise, y_noise, test_size=0.3, random_state=42)

# Fit a linear regression model on all datasets
model = LinearRegression()

# Evaluate R2 Score for each dataset
r2_simple = model.fit(X_train_simple, y_train_simple).score(X_test_simple, y_test_simple)
r2_non_linear = model.fit(X_train_non_linear, y_train_non_linear).score(X_test_non_linear, y_test_non_linear)
r2_noise = model.fit(X_train_noise, y_train_noise).score(X_test_noise, y_test_noise)

# Print the R2 scores
print(f"R2 Score for Simple Linear Data: {r2_simple:.2f}")
print(f"R2 Score for Non-linear Data: {r2_non_linear:.2f}")
print(f"R2 Score for Noisy Data: {r2_noise:.2f}")

# Plotting to visualize the data
plt.figure(figsize=(14, 8))

plt.subplot(1, 3, 1)
plt.scatter(X_test_simple, y_test_simple, color='blue', label='True Values')
plt.plot(X_test_simple, model.predict(X_test_simple), color='red', label='Predictions')
plt.title(f'Simple Linear Data - R2: {r2_simple:.2f}')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

plt.subplot(1, 3, 2)
plt.scatter(X_test_non_linear, y_test_non_linear, color='blue', label='True Values')
plt.plot(X_test_non_linear, model.predict(X_test_non_linear), color='red', label='Predictions')
plt.title(f'Non-linear Data - R2: {r2_non_linear:.2f}')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

plt.subplot(1, 3, 3)
plt.scatter(X_test_noise, y_test_noise, color='blue', label='True Values')
plt.plot(X_test_noise, model.predict(X_test_noise), color='red', label='Predictions')
plt.title(f'Noisy Data - R2: {r2_noise:.2f}')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```

Task 3: Use a sample dataset, compute all three metrics, and deduce model performance.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Task 3: Use a sample dataset, compute all three metrics, and deduce model performance.

# Sample dataset: Predicting house prices based on the size of the house
np.random.seed(42)
X = np.random.rand(100, 1) * 2000  # Size of the house in square feet (1 to 2000)
y = 50 * X.ravel() + 200000 + np.random.normal(0, 10000, 100)  # Price in dollars with some noise

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate all three metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the metrics
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R2 Score: {r2:.2f}")

# Visualizing the predictions vs true values
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='True Values', alpha=0.7)
plt.plot(X_test, y_pred, color='red', label='Predictions')
plt.title(f'Model Performance: MAE = {mae:.2f}, MSE = {mse:.2f}, R2 = {r2:.2f}')
plt.xlabel('Size of House (sq. ft.)')
plt.ylabel('Price of House (USD)')
plt.legend()
plt.show()
```
