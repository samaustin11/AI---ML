Question 8: Logistic Regression with L1 Regularization<br>
Task: Implement logistic regression with L1 regularization and observe the effect on feature selection.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Task: Implement logistic regression with L1 regularization and observe the effect on feature selection.

# Generate a synthetic classification dataset
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_classes=2, random_state=42)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression with L1 Regularization (Lasso)
log_reg_l1 = LogisticRegression(penalty='l1', solver='liblinear', random_state=42)
log_reg_l1.fit(X_train, y_train)

# Predict and evaluate the model
y_pred = log_reg_l1.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy with L1 Regularization: {accuracy:.2f}")

# Get the coefficients of the model to observe feature selection
coefficients = log_reg_l1.coef_.ravel()

# Visualize the coefficients to see which features are selected (non-zero coefficients)
plt.figure(figsize=(10, 6))
plt.bar(range(len(coefficients)), coefficients)
plt.title('Feature Coefficients with L1 Regularization (Logistic Regression)')
plt.xlabel('Feature Index')
plt.ylabel('Coefficient Value')
plt.show()

# Print the selected features (non-zero coefficients)
selected_features = np.where(coefficients != 0)[0]
print(f"Selected Features: {selected_features}")
```
