Question 1: Understanding the Dataset 
<br>
Description: Load a dataset and understand its basic properties including data types dimensions, and first few rows

```python
import pandas as pd

# Create a sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [23, 45, 34, 25, 56],
            'Price': [1000, 2500, 1500, 1200, 3000],
                'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
                }
# Load the dataset into a DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataset
print(df.head())

# Display the dimensions of the dataset (rows, columns)
print(df.shape)

# Display the data types of each column
print(df.dtypes)
```

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()

# Display the number of missing values in each column
print(missing_values)
```


Question 5: Categorical Data Analysis
<br>
Description: Explore the counts of categorical variables.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [23, 45, 34, 25, 56],
            'Price': [1000, 2500, 1500, 1200, 3000],
                'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
                }

# Load the data into a DataFrame
df = pd.DataFrame(data)

# Task: Categorical Data Analysis (Exploring counts of 'City' variable)
category_counts = df['City'].value_counts()

# Display the count of unique categories
print(category_counts)

# Plotting the bar chart
category_counts.plot(kind='bar')

# Show the plot
plt.title('City Count Distribution')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

Question 6: Data Transformation
<br>
Description: Transform a categorical column into numerical using Label Encoding.

```python
from sklearn.preprocessing import LabelEncoder

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [23, 45, 34, 25, 56],
            'Price': [1000, 2500, 1500, 1200, 3000],
                'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
                }

df = pd.DataFrame(data)

# Task: Transform 'City' column into numerical using Label Encoding
label_encoder = LabelEncoder()

# Apply label encoding to 'City' column
df['City_Encoded'] = label_encoder.fit_transform(df['City'])

# Display the DataFrame with the new encoded column
print(df)
```

Question 8: Correlation Analysis
<br>
Description: Calculate and visualize the correlation matrix for numerical features.

```python
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = df[['Age', 'Price']].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

Question 9: Feature Engineering
<br>
Description: Create a new feature by combining or transforming existing features.

```python
# Create a new feature 'Age_Price_Ratio' by dividing Age by Price
df['Age_Price_Ratio'] = df['Age'] / df['Price']

# Display the first few rows of the dataset to see the new feature
print(df.head())
```

Question 10: Advanced Outlier Detection
<br>
Description: Use the Z-score method to identify and handle outliers.

```python
from scipy import stats
import numpy as np

# Calculate the Z-scores for the numerical columns
z_scores = np.abs(stats.zscore(df[['Age', 'Price']]))

# Set a threshold for Z-score (commonly used threshold is 3)
threshold = 3

# Identify outliers where the Z-score is greater than the threshold
outliers = (z_scores > threshold).all(axis=1)

# Filter out the outliers
df_no_outliers = df[~outliers]

# Display the dataset without outliers
print(df_no_outliers.head())
```
