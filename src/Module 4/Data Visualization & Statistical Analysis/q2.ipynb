Title: Correlation Heatmaps<br>

Task 1: Generate a correlation heatmap of the dataset.<br>
Task 2: Highlight correlation between sepal length and petal length.<br>
Task 3: Highlight correlation between petal width and petal length.<br>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Exclude non-numeric columns for correlation calculation
numeric_data = iris.select_dtypes(include=['float64', 'int64'])

# Task 1: Generate a correlation heatmap of the dataset
plt.figure(figsize=(8, 6))
corr = numeric_data.corr()  # Compute correlation only for numeric features
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()

# Task 2: Highlight correlation between sepal length and petal length
plt.figure(figsize=(8, 6))
sns.heatmap(corr[['sepal_length', 'petal_length']].loc[['sepal_length', 'petal_length']], annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation between Sepal Length and Petal Length')
plt.show()

# Task 3: Highlight correlation between petal width and petal length
plt.figure(figsize=(8, 6))
sns.heatmap(corr[['petal_width', 'petal_length']].loc[['petal_width', 'petal_length']], annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation between Petal Width and Petal Length')
plt.show()
```

Title: Principal Component Analysis (PCA)

Task 1: Perform PCA and plot the first two principal components.
Task 2: Visualize explained variance by each component.
Task 3: Retain more components and visualize in 3D.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Task 1: Perform PCA and plot the first two principal components
# Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(iris.drop(columns='species'))

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Plot the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=iris['species'].map({'setosa': 0, 'versicolor': 1, 'virginica': 2}), cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - First Two Principal Components')
plt.colorbar(label='Species')
plt.show()

# Task 2: Visualize explained variance by each component
explained_variance = pca.explained_variance_ratio_

# Plot explained variance
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance)
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Each Principal Component')
plt.show()

# Task 3: Retain more components and visualize in 3D
# Perform PCA with 3 components
pca_3d = PCA(n_components=3)
pca_result_3d = pca_3d.fit_transform(scaled_data)

# 3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the first three principal components
ax.scatter(pca_result_3d[:, 0], pca_result_3d[:, 1], pca_result_3d[:, 2], c=iris['species'].map({'setosa': 0, 'versicolor': 1, 'virginica': 2}), cmap='viridis')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('PCA - First Three Principal Components')
plt.show()




```

**Statistical Analysis in EDA**<br>

Objective: Calculate basic statistical metrics and explore the relationship between features using correlation and covariance.<br>

Title: Descriptive Statistics<br>

Task 1: Calculate mean, median, and standard deviation of petal length.<br>
Task 2: Calculate skewness and kurtosis of sepal width.<br>
Task 3: Calculate mean, median, and standard deviation of sepal length.

```python
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Task 1: Calculate mean, median, and standard deviation of petal length
petal_length = iris['petal_length']

mean_petal_length = np.mean(petal_length)
median_petal_length = np.median(petal_length)
std_petal_length = np.std(petal_length)

print(f"Petal Length - Mean: {mean_petal_length}, Median: {median_petal_length}, Standard Deviation: {std_petal_length}")

# Task 2: Calculate skewness and kurtosis of sepal width
sepal_width = iris['sepal_width']

sepal_width_skewness = skew(sepal_width)
sepal_width_kurtosis = kurtosis(sepal_width)

print(f"Sepal Width - Skewness: {sepal_width_skewness}, Kurtosis: {sepal_width_kurtosis}")

# Task 3: Calculate mean, median, and standard deviation of sepal length
sepal_length = iris['sepal_length']

mean_sepal_length = np.mean(sepal_length)
median_sepal_length = np.median(sepal_length)
std_sepal_length = np.std(sepal_length)

print(f"Sepal Length - Mean: {mean_sepal_length}, Median: {median_sepal_length}, Standard Deviation: {std_sepal_length}")
```
